```python
mis_tokens={}
def tokenizar(dato):
    token=str(uuid.uuid4())
    mis_tokens[token] =dato
    return token

def recuperar_dato(token):
    return mis_tokens.get(token, "token no valido")

dato_original = "123-456-789"
token=tokenizar(dato_original)
print(f"token generado{token}")
print(f"dato recuperado: {recuperar_dato(token)}")
```

```python
d=pd.DataFrame({
    "edad":[22,45,36,29,58],
    "genero":["f","m","m","f","m"],
    "contratado":[1,0,1,0,1]
})
d
```

```python
agrupado = d.groupby("genero")
for nombre, grupo in agrupado:
    print(nombre, grupo)
```

```python
data_balanceadas=pd.DataFrame()
for nombre, grupo in  agrupado:
    grupo_balanceado=resample(grupo,replace=True,n_samples =10,random_state=123)
    data_balanceadas =pd.concat([data_balanceadas,grupo_balanceado])
data_balanceadas
```
