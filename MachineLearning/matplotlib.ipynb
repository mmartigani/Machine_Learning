```python
dd=pd.read_csv("Datos+Meteorológicos_Arg_2023.csv")
dd.head(2)
```

```python
dd["Fecha"]=pd.to_datetime(dd["Fecha"],format='%d/%m/%Y')
dd.head(1)
```

```python
lista_ciudades = []
for c in dd["Ciudad"]:
    if c not in lista_ciudades:
        lista_ciudades.append(c)
lista_ciudades
```

```python
dict_meses = {1:"enero",
             2:"febrero",
             3:"marzo",
             4:"abril",
             5:"mayo",
             6:"junio",
             7:"julio",
             8:"agosto",
             9:"septiembre",
             10:"octubre",
             11:"noviembre",
             12:"diciembre"}
```

```python
def consultar_temperaturas():
    while True:
        #solicitar la ciudad
        print("\nCuidades disponibles",lista_ciudades)
        ciudad_elegida=input("elija la ciudad de la lista: ").title()
        
        #solicitar el mes
        mes_elegido =int(input("elija un mes: ejemplo enero es 1: "))
        #validar los inputs 
        if ciudad_elegida not in lista_ciudades or mes_elegido not in range(1,13):
            print("Ciudad o mes no valido")
            continue
        #crear un dataframe 
        datos_ciudad_mes = dd[(dd["Ciudad"]== ciudad_elegida) &(dd["Fecha"].dt.month == mes_elegido)]
        
        #graficamos las temperaturas 
        plt.figure(figsize=(10,6))
        plt.plot(datos_ciudad_mes["Fecha"], datos_ciudad_mes["Temperatura Maxima"],label="Maxima", color="red")
        plt.plot(datos_ciudad_mes["Fecha"], datos_ciudad_mes["Temperatura Minima"],label="Minima", color="blue")
        plt.title(f"temperaturas {ciudad_elegida}, durante el mes {dict_meses[mes_elegido]}")
        plt.xlabel("Fecha")
        plt.ylabel("Temperaturas")
        plt.legend()
        plt.xticks(rotation=45)
        plt.show()
       #preguntar al user si quiere continuar
        otra=input("deseas continuar consultando (s/n): ")
        if otra.lower() != 's':
            break
```

```python
consultar_temperaturas()
```
