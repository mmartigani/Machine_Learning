```python
import pandas as pd 
import numpy as np 
import time
import matplotlib.pyplot as plt
%matplotlib inline 
#permite que le coloquemos ; en vez de show()
import seaborn as sns 
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
r=pd.read_csv("datos_seguro.csv")
r.head(5)
```

```python
plt.scatter(r.edad, r.compra);
```

```python
X_entrena, X_prueba,y_entrena, y_prueba= train_test_split(r[["edad"]],r["compra"],
                                                         train_size=0.9);
```

```python
X_entrena.head(5)
```

```python
X_prueba
```

```python
modelo = LogisticRegression()
```

```python
modelo.fit(X_entrena, y_entrena)
```

```python
modelo.score(X_entrena, y_entrena)
```

```python
datos_nuevos =pd.DataFrame({"edad":[25,35,45,55]})
datos_nuevos
```

```python
probabilidades = modelo.predict_proba(datos_nuevos)
probabilidades
```

```python
prob_compra = probabilidades[:,1]
prob_compra
#array con los 4 indices de probabilidades
```

```python
plt.scatter(r.edad, r.compra)
plt.scatter(datos_nuevos["edad"], prob_compra,color="red");
```
