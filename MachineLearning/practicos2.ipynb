```python
import pandas as pd 
import numpy as np 
import time
import matplotlib.pyplot as plt
%matplotlib inline 
#permite que le coloquemos ; en vez de show()
import seaborn as sns 
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression , LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.cluster import KMeans
```

```python
print("algoritmo KMEANS CLUSTERING BUSCA DE CENTROIDES") 
```

```python
#inventamos una base de datos 
np.random.seed(42)
data = np.random.rand(100,2)
data
```

```python
#tengo la variable y la entrenamos en clusters con su centroide
kmeans = KMeans(n_clusters=5,random_state=42)
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
centroides
```

```python
etiquetas = kmeans.labels_
etiquetas
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data[:,0],data[:,1],c=etiquetas, cmap="viridis", marker="o")
plt.scatter(centroides[:,0],centroides[:,1],c="red",marker="x",s=200,label="centroides")
plt.title("Visualizacion de lmeans clustering")
plt.legend()
```

```python
pinguinos = sns.load_dataset('penguins')
```

```python
pinguinos.dropna(inplace=True) #nos aseguramos que no haya valores nulos 
```

```python
pinguinos.head(1)
```

```python
data = pinguinos[['bill_length_mm','bill_depth_mm']]
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
etiquetas =kmeans.labels_
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data=pinguinos,x='bill_length_mm', y='bill_depth_mm', hue=etiquetas,palette='viridis')
plt.scatter(centroides[:,0],centroides[:,1],c="red",s=100,marker="x",label="centroides")
plt.title("kmeans clustering en el conjunto de pinguinos")
plt.legend()
plt.xlabel("longuitud del pico(mm)")
plt.ylabel("profundidad del pico (mm)");
```

```python
import pandas as pd 
import numpy as np 
import time
import matplotlib.pyplot as plt
%matplotlib inline 
#permite que le coloquemos ; en vez de show()
import seaborn as sns 
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression , LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
iris = load_iris()
```

```python
X=iris.data
y=iris.target
```

```python
#estandarizacion de datos TODOS TIENEN UN PUNTO DE PARTIDA COMUN
#NOS AYUDA A SABER QUIEN SOBRESALE 
X_centrado = X-np.mean(X,axis=0)
```

```python
pca =PCA(n_components=2) #COLOCO 2 COMPONENTES PRINCIPALES
```

```python
X_pca =pca.fit_transform(X_centrado)
```

```python
especies = ["setosa","versicolor", "virginica"]
```

```python
plt.figure(figsize=(8,6))
for i in range(0,3):
    plt.scatter(X_pca[y==i,0],X_pca[y==i,1],label=especies[i])
plt.xlabel("primer comoponente principal")
plt.ylabel("segundo componente principal")
plt.legend()
plt.title("PCA DEL CONJUNTO IRIS");
```
